
# 🔍 Analyse d'Explicabilité du Modèle — SHAP
---

## 📊 Paragraphe 1 — Interprétation du premier graphique (Feature Importance)

> **Quelles variables influencent le plus les prédictions ?**

Le premier graphique représente l'importance globale des variables dans le modèle à l'aide des valeurs SHAP. Il permet d'identifier quelles caractéristiques influencent le plus la prédiction du niveau d'obésité.

On observe que **Weight (poids)** est la variable la plus importante, suivie par **Height (taille)**, **Gender (genre)** et **Age (âge)**. Cela signifie que le modèle utilise principalement ces facteurs pour prédire le niveau d'obésité d'un individu.

Ce résultat est cohérent avec les connaissances médicales, car le poids et la taille sont directement liés à l'**indice de masse corporelle (IMC)**, un indicateur clé pour évaluer l'obésité. Les autres variables, comme les habitudes alimentaires ou le mode de vie, ont une influence plus faible mais contribuent tout de même à améliorer la précision du modèle.

##  Paragraphe 2 — Interprétation du second graphique (SHAP Summary Plot)

> **Comment chaque variable impacte-t-elle individuellement les prédictions ?**

Le second graphique est un **SHAP summary plot** qui montre l'impact de chaque variable sur les prédictions du modèle pour chaque individu du jeu de données.

| Élément | Signification |
|--------|--------------|
| 🔵 Point bleu | Valeur faible de la variable |
| 🔴 Point rouge | Valeur élevée de la variable |
| ← Position gauche | Diminue la probabilité d'obésité |
| → Position droite | Augmente la probabilité d'obésité |

Par exemple, des valeurs élevées de **Weight** tendent à augmenter la probabilité d'obésité, tandis que certaines variables liées au mode de vie, comme la consommation de légumes **(FCVC)**, peuvent contribuer à réduire ce risque.

Ce graphique permet donc de comprendre non seulement **quelles variables sont importantes**, mais aussi **comment elles influencent concrètement** les prédictions du modèle.
